```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def init_param():
    k = 1.3
    w = 0.4
    params = k, w
    return params

def map(x, k, w):
    val = x + w - k / ( 2 * np.pi) * np.sin(2 * np.pi * x)
    if val > 1: return val - 1
    if val < -1: return val + 1
    return val

def main():
    plt.figure(figsize=(5,5))
    k, w = init_param()
    kk = np.linspace(0.6,2,5)
    kk = [0.1, 0.6, 0.95, 1.3, 1.65, 2.1]
    n = 10000
    xx = np.zeros(n)
    yy = xx
    for k in kk:
        for i in range(n-1):
            yy[i+1] = map(xx[i], k, w)
        plt.scatter(xx[:-1], yy[1:], s = 1, alpha=0.1, label = k)
    plt.plot(np.linspace(0,1,10), np.ones(10)*w, linestyle='--', linewidth=0.9)
    plt.xlabel(r'$x_{n}$', fontsize=20)
    plt.ylabel(r'$x_{n+1}$', fontsize=20)
    leg = plt.legend(markerscale=4)
    for lh in leg.legendHandles: 
        lh.set_alpha(1)
    plt.xlim(0,1)
    plt.ylim(0,1)
    plt.savefig('../LaTex/figures/4_3_py.png', dpi = 300)
    
def main1():
    plt.figure(figsize=(5,5))
    k, w = init_param()
    kk = np.linspace(0.6,2,5)
    kk = [0.6, 0.95, 1.3, 1.65, 2.1]
    n = 10000
    xx = np.zeros(n)
    yy = xx
    for k in kk:
        for i in range(n-1):
            yy[i+1] = map(xx[i], k, w)
        plt.scatter(xx[:-1], yy[1:], s = 1, alpha=0.1, label = k)
    plt.plot(np.linspace(0,1,10), np.ones(10)*w, linestyle='--', linewidth=0.9)
    plt.xlabel(r'$x_{n}$')
    plt.ylabel(r'$x_{n+1}$')
    leg = plt.legend(markerscale=4)
    for lh in leg.legendHandles: 
        lh.set_alpha(1)
    plt.xlim(0,1)
    plt.ylim(0,1)
    plt.savefig('../LaTex/figures/4_3_py.png', dpi = 300)
main()
```
